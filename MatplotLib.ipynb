```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
a=np.linspace(0,180,13)
x=np.sin(a)
y=np.cos(a)

print(a)
print(x)
print(y)
```

ploting a and x

```python
plt.plot(a,x,'r-')
plt.show()
```

PUTTING LABELS

```python
plt.plot(a,x,'r-')
plt.xlabel('angles')
plt.ylabel('sine')
plt.show()
```

TWO FIGURE IN ONE

```python
plt.plot(a,x,'r-')
plt.plot(a,y,'g--')
plt.xlabel('angles')
plt.ylabel('sine')
plt.show()
```

PLotting Bar

```python
fig=plt.figure()
fx=fig.add_axes([0,0,1,1])
fruits=['apple','mango','banana','cherry']
price=[10,20,30,40]

fx.bar(fruits,price)
plt.xlabel('fruits')
plt.ylabel('price')
plt.title('PRICE OF FRUIT')
plt.show()
```

PIE CHART

```python
fig=plt.figure()
fx=fig.add_axes([0,0,1,1])
fruits=['apple','mango','banana','cherry']
price=[10,20,30,40]
fx.pie(price,labels=fruits,autopct='%1f%%')
plt.show()
```

3D scattering

```python
a=np.linspace(0,180,13)
x=np.sin(a)
y=np.cos(a)

figScatter=plt.figure()
fx=plt.axes(projection='3d')
fx.scatter(x,y,a,c=a,cmap='Blues')

plt.show()
```

SCATTER PLOT

```python
a=np.linspace(0,180,13)
x=np.sin(a)
y=np.cos(a)

figScatter=plt.figure()
fx=figScatter.add_axes([0,0,1,1])
fx.scatter(x,a,color='g')
fx.scatter(y,a,color='r')

plt.show()
```
